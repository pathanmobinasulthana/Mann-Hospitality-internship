### Creating a function that takes dataseta as input  and  return the cleaned and formated data according to our requirement.

```python

import re
import datetime

def extract_text_from_values(value_list):
    extracted_values = []
    pattern = r'(_Kg_Qty|_Pcs_Qty|_Rate|_Amount|__Qty)'  # Regular expression pattern for matching '_qnty', '_ate', or '_amount'

    for value in value_list:
        match = re.search(pattern, value)

        if match:
            # Extract the prefix before the match
            prefix = value[:match.start()]

            extracted_values.append(prefix)
        else:
            extracted_values.append(value)

    return extracted_values

def Mannhospitality(df):

    l = []
    for i in df.iloc[7]:
        l.append(i)

    d = df[((df[1] == datetime.datetime(2023, 6, 6, 0, 0)) | (df[1] ==datetime.datetime(2023, 6, 7, 0, 0)))]
    d.columns = l
    d = d[['Bill Date', 'Item Name', 'Vendor Name','FF OUT_Kg_Qty',   'FF OUT_Pcs_Qty',       'GLL_Kg_Qty',
            'GLL_Pcs_Qty',       'AFF_Kg_Qty',      'AFF_Pcs_Qty',
            'SKIN OUT_Kg_Qty', 'SKIN OUT_Pcs_Qty',       'TDR_Kg_Qty',
            'TDR_Pcs_Qty',       'BRR_Kg_Qty',      'BRR_Pcs_Qty',
            'CRR_Kg_Qty',      'CRR_Pcs_Qty',       'KBB_Kg_Qty',
            'KBB_Pcs_Qty',       'KLL_Kg_Qty',      'KLL_Pcs_Qty',
            'LPP_Kg_Qty',      'LPP_Pcs_Qty',       'WNG_Kg_Qty',
            'WNG_Pcs_Qty',       'SWW_Kg_Qty',       'CBL_Kg_Qty',
            'TBL_Kg_Qty',         'OTH__Qty',           'Border',
            '_FF OUT_Rate',        '_GLL_Rate',        '_AFF_Rate',
            '_SKIN OUT_Rate',        '_TDR_Rate',        '_BRR_Rate',
            '_CRR_Rate',        '_KBB_Rate',        '_KLL_Rate',
            '_LPP_Rate',        '_WNG_Rate',        '_SWW_Rate',
            '_CBL_Rate',        '_TBL_Rate',        '_OTH_Rate',
            'Border',   '_FF OUT_Amount',      '_GLL_Amount',
            '_AFF_Amount', '_SKIN OUT_Amount',      '_TDR_Amount',
            '_BRR_Amount',      '_CRR_Amount',      '_KBB_Amount',
            '_KLL_Amount',      '_LPP_Amount',      '_WNG_Amount',
            '_SWW_Amount',      '_CBL_Amount',      '_TBL_Amount',
            '_OTH_Amount',]]

    i = d.melt(id_vars=["Bill Date", 'Vendor Name'])
    i = i[(i["variable"] != 'Item Name')]

    if 'Syed' in df[2].unique():
        i['Vendor Name'] = "Syed"
    elif "SK" in df[2].unique():
        i['Vendor Name'] = "Sk"
    else:
        i['Vendor Name'] ="BSML"
    i = i[i["variable"] != "Item Name"]

    # Create new columns for Qty, Pcs, and Amount and initialize them with NaN
    i["Qty"] = pd.NA
    i["Rate"] = pd.NA
    i["Amount"] = pd.NA
    i["Pcs"] = pd.NA

    # Define the conditions to extract values based on text patterns
    qnty_condition = i["variable"].str.contains("_Kg_Qty")
    pcs_condition = i["variable"].str.contains("_Pcs_Qty")
    amount_condition = i["variable"].str.contains("_Amount")
    rate_condition = i["variable"].str.contains("_Rate")

    # Extract values based on conditions and update the corresponding columns
    i.loc[qnty_condition, "Qty"] = i.loc[qnty_condition, "value"]
    i.loc[pcs_condition, "Pcs"] = i.loc[pcs_condition, "value"]
    i.loc[amount_condition, "Amount"] = i.loc[amount_condition, "value"]
    i.loc[rate_condition, "Rate"] = i.loc[rate_condition, "value"]

    # Drop the original "value" column
    i.drop(columns=["value"], inplace=True)

    if 'Syed' in i["Vendor Name"].unique():
        w = i[(i["variable"] == 'SKIN OUT') | (i["variable"] == 'BRR') | (i["variable"] == 'WNG') | (i["variable"] == 'SWW') | (i["variable"] == 'FF OUT') | (i["variable"] == 'CBL')]
    elif 'SK' in i["Vendor Name"].unique():
        w = i[(i["variable"] == 'KLL') | (i["variable"] == 'TDR') | (i["variable"] == 'WNG') | (i["variable"] == 'SWW') | (i["variable"] == 'FF OUT') | (i["variable"] == 'CBL') | (i["variable"] == 'TBL')]
    else:
        w = i[(i["variable"] == 'KLL') | (i["variable"] == 'TDR') | (i["variable"] == 'WNG') | (i["variable"] == 'SWW') | (i["variable"] == 'FF OUT') | (i["variable"] == 'CBL') | (i["variable"] == 'TBL') | (i["variable"] == 'SKIN OUT')]

    i.fillna(0.0, inplace=True)
    l=extract_text_from_values(i["variable"])
    l=[i.replace("_","" )  for i in l]
    i['Item Name']=l
    i[['Qty', 'Rate', 'Amount', 'Pcs']] = i[['Qty', 'Rate', 'Amount', 'Pcs']].astype(float)
    aggregation = {
        'Qty': 'sum',
        'Rate': 'max',
        'Amount': 'sum',
        'Pcs': 'max'
    }
    result = i.groupby(['Bill Date', 'Vendor Name', 'Item Name']).agg(aggregation).reset_index()
    result = result[result["Qty"] != 0.0]

    return result
```
